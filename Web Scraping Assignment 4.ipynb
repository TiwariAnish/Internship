2. Scrape the details team India’s international fixtures from bcci.tv.   
Url = https://www.bcci.tv/.   
You need to find following details:   
A) Series   
B) Place   
C) Date   
D) Time   
Note: - From bcci.tv home page you have reach to the international fixture page through code. 

```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
import warnings 
warnings.filterwarnings('ignore')
import pandas as pd
import time
```

```python
# Initializing the driver

driver = webdriver.Chrome()
```

```python
# Browsing the website

driver.get('https://www.bcci.tv/')
```

```python
# Clicking on the fixtures button on web page

fixture_click = driver.find_element(By.XPATH, '//div[@class="imw-tabs international-tabs"]/a[2]')
fixture_click.click()
```

```python
# scraping the Series, Place, Date and Time information

Series = []
Place = []
Date = []
Time = []

start = 0
end = 5

for page in range(start, end):
    series_tag = driver.find_elements(By.XPATH, '//div[@class="match-info"]/h5')
    for i in series_tag:
        Series.append(i.text)
        
    place_tag = driver.find_elements(By.XPATH, '//div[@class="match-venue ng-scope"]')
    for i in place_tag:
        Place.append(i.text)
        
    date_tag = driver.find_elements(By.XPATH, '//div[@class="match-dates ng-binding"]')
    for i in date_tag:
        Date.append(i.text)
        
    time_tag = driver.find_elements(By.XPATH, '//div[@class="match-time no-margin ng-binding"]')
    for i in time_tag:
        Time.append(i.text)
        
    more_matches_button = driver.find_element(By.XPATH, '//div[@class="col-lg-12 col-md-12 col-sm-12"]/button')
    more_matches_button.click()
    time.sleep(3)
```

```python
# Watching the length of the each elements

print(len(Series))
print(len(Time))
print(len(Place))
print(len(Date))
```

```python
# Creating DataFrame for storing the scraped data

df = pd.DataFrame({'Series Name': Series, 'Place/Venue Name': Place, 'Date': Date, 'Time of Match': Time})
df
```

4. Scrape the details of trending repositories on Github.com.

B) Repository description   
C) Contributors count   
D) Language used   

```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import ElementNotInteractableException
import warnings 
warnings.filterwarnings('ignore')
import time
import pandas as pd
```

```python
# Initializing the driver

driver = webdriver.Chrome()
```

```python
# Browsing the website
driver.get('https://github.com/')
```

```python
# clicking on Trending Repositories
try:
    trending = driver.find_element(By.XPATH, '/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[4]/div/div[3]/ul/li[2]/a')
    trending.click()
except ElementNotInteractableException as e:
    print('Exception Raised: ',e)
    trending1 = driver.find_element(By.XPATH, '/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[4]/div/div[3]/ul/li[2]/a')
    driver.get(trending1.get_attribute('href'))
    
```

```python
# Scraping the Repository Title

Repo_title = []

title_tag = driver.find_elements(By.XPATH, '//h2[@class="h3 lh-condensed"]')
for i in title_tag:
    title = i.text
    Repo_title.append(title)
    
print(len(Repo_title))
print(Repo_title)
```

```python
# Scraping the Repository Description

Repo_description = []

description_tag = driver.find_elements(By.XPATH, '//p[@class="col-9 color-fg-muted my-1 pr-4"]')
for i in description_tag:
    description = i.text
    Repo_description.append(description)
    
print(len(Repo_description)) 
print(Repo_description)
```

```python
# Scraping the Repository URLs

Repo_urls = []

start = 0
end =1

for page in range(start, end):
    url = driver.find_elements(By.XPATH, '//h2[@class="h3 lh-condensed"]/a')
    for i in url:
        Repo_urls.append(i.get_attribute('href'))
print(len(Repo_urls))
print(Repo_urls)
```

```python
# Clicking a repository for scraping the in detail and get better results for Contributors Count


click_repo = driver.find_element(By.XPATH, '//h2[@class="h3 lh-condensed"]/a')
driver.get(click_repo.get_attribute('href'))
```

```python
# Scraping the Repository contributors

totalContributor = []

for i in Repo_urls[:15]:
    driver.get(i)
    time.sleep(3)
    
    contributor = driver.find_elements(By.XPATH, '//a[@class="Link--primary no-underline Link d-flex flex-items-center"]')
    for contribute in contributor[:15]:
        totalContributor.append(contribute.text)
    
print(len(totalContributor))
print(totalContributor)
```

```python
# Extracting the contributors count

totalContributors = [int(item.split('\n')[1].replace(',', '')) for item in totalContributor if 'Contributors' in item]
print(len(totalContributors))
print(totalContributors)
```

```python
# Scraping the Repository Language used

Lang_used = []

lang_tag = driver.find_elements(By.XPATH, '//span[@itemprop="programmingLanguage"]')
for i in lang_tag:
    language = i.text
    Lang_used.append(language)
    
print(len(Lang_used))
print(Lang_used)
```

```python
# Creating the DataFrame of Scraped data

df = pd.DataFrame({'Repostory Title': Repo_title, 'Repository Description': Repo_description, 'Contributor Count': totalContributors, 'Language Used': Lang_used})
df
```

```python
# Scraping the required data

Dataset_name = []
Data_type = []
Task = []
Attribute_type = []
Num_instance = []
Num_attribute = []
Year = []

for i in Dataset_urls:
    driver.get(i)
    time.sleep(5)
    
    name_tag = driver.find_elements(By.XPATH, '//div[@class="flex flex-col"]/div/div/div/h1')
    for k in range(0, len(name_tag), 7):
        name = name_tag[k].text
        Dataset_name.append(name)
    
    type_tag = driver.find_elements(By.XPATH, '//div[@class="col-span-4"]/p')
    for k in range(0, len(type_tag), 6):
        type = type_tag[k].text
        Data_type.append(type)
    
    task_tag = driver.find_elements(By.XPATH, '//div[@class="col-span-4"]/p')
    for k in range(2, len(task_tag), 4):
        task = task_tag[k].text
        Task.append(task)
        
    attr_type_tag = driver.find_elements(By.XPATH, '//div[@class="col-span-4"]/p')
    for k in range(3, len(attr_type_tag), 4):
        attr_type = attr_type_tag[k].text
        Attribute_type.append(attr_type)
    
    num_instance_tag = driver.find_elements(By.XPATH, '//div[@class="col-span-4"]/p')
    for k in range(4, len(num_instance_tag), 4):
        instance = num_instance_tag[k].text
        Num_instance.append(instance)
        
    num_attribute_tag = driver.find_elements(By.XPATH, '//div[@class="col-span-4"]/p')
    for k in range(5, len(num_attribute_tag), 4):
        num_attri = num_attribute_tag[k].text
        Num_attribute.append(num_attri)
        
print(len(Dataset_name))
print(Dataset_name)

print(len(Data_type))
print(Data_type)

print(len(Task))
print(Task)

print(len(Attribute_type))
print(Attribute_type)

print(len(Num_instance))
print(Num_instance)

print(len(Num_attribute))
print(Num_attribute)

```
